<h2>Data Visualization</h2>

<h3> Histplot of total mark</h3>
<h3>Histplot of averagemark with respected to gender</h3>

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=df, x="Total_score", kde=True, bins=30)

plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Total_score", hue="gender", kde=True, bins=30)

plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=df, x="Average_score", kde=True, bins=30)

plt.subplot(1, 2, 2)
sns.histplot(data=df, x="Average_score", hue="gender", kde=True, bins=30)

plt.show()
```
